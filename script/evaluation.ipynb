```python
import json
```

```python
CPC_Mapping = {}
```

```python
with open('../data/output/llama-3-70b-chat/cont_generation_1.jsonl', 'r') as file:
    # Iterate through each line
    for line in file:
        # Load the JSON object from the line
        data = json.loads(line)
        
        # Process the JSON object
        # For example, you can access specific fields
        cpc_codes = data['cpc_codes']
        topic_names = set([topic["name"] for topic in data["topics"]])
        y02_codes = set([code.split()[0] for code in cpc_codes if code.startswith('Y02')])
        print(y02_codes)
        print(topic_names)
        for y02_code in y02_codes:
            if y02_code not in CPC_Mapping:
                CPC_Mapping[y02_code] = set(topic_names)
            else:
                CPC_Mapping[y02_code].update(topic_names)
        
        
```

```python
for cpc_code in CPC_Mapping:
    print(cpc_code)
    print(len(CPC_Mapping[cpc_code]))
```
